```python
foo = []
for j in range(1,16): 
    mylist=list(votes.iloc[0:6,0])
    myorder=[x - 1 for x in list(votes.iloc[0:6,j].astype(int)) ]
    mylist = [ mylist[i] for i in myorder]
    mylist = [[y] for y in mylist]
    foo.append({"count":1,"ballot": mylist})
```

```python
IRV(foo, ballot_notation = CondorcetHelper.BALLOT_NOTATION_GROUPING).as_dict()
```

```python
SchulzeSTV(foo, ballot_notation = CondorcetHelper.BALLOT_NOTATION_GROUPING).as_dict()['winners']
```

```python
SchulzeMethod(foo, ballot_notation = CondorcetHelper.BALLOT_NOTATION_GROUPING).as_dict()['winner']
```
